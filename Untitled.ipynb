```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
```

# Read myopia csv into Dataframe

```python
# create path to csv
file_path = 'myopia.csv'

# variable that reads csv file
df = pd.read_csv(file_path)

df.sample(10)
```

# Drop columns that are not needed from dataset

```python
# drop myopia column
no_myo = df.drop(['MYOPIC','DIOPTERHR'], axis='columns')
labels = df['MYOPIC']
```

```python
no_myo
```

# Standardize my dataset

```python
# Create a scaler to standardize the data
scaler = StandardScaler()
```

```python
# Split the dataframe into data and target
y = df['MYOPIC'].values
x = no_myo
```

```python
MYO_scaled = StandardScaler().fit_transform(x)
```

```python
MYO_scaled
```

```python
#change to df to be used in PCA reduction
MYO_scaled_df = pd.DataFrame(MYO_scaled)
MYO_scaled_df.head()
```

# Perform dimensionality reduction with PCA

```python
# before reduction
MYO_scaled.shape
```

```python
# PCA model to reflect 90% variance
pca = PCA(n_components = .90)

# use fit transform on pca model
MYO_morph = pca.fit_transform(MYO_scaled)
```

```python
# view number of principal components
MYO_morph.shape
```

```python
# convert to df to use in scatter plot
MYO_reduc_df = pd.DataFrame(MYO_morph)
MYO_reduc_df.head()
```

The dataset has gone from 13 columns to 10!

# Reducing Further with t-SNE

```python
# initialize t_SNE model
tsne = TSNE(learning_rate=35, perplexity = 50)
```

```python
# Reduce dimensions
tsne_features = tsne.fit_transform(MYO_reduc_df)
```

```python
tsne_features.shape
```

# create scatter plot

```python

# The first column of transformed features
MYO_reduc_df['x'] = tsne_features[:,0]

# The second column of transformed features
MYO_reduc_df['y'] = tsne_features[:,1]
```

```python
MYO_reduc_df.shape
```

```python
# visualize clusters
plt.scatter(MYO_reduc_df['x'], MYO_reduc_df['y'])
plt.show()
```

```python
# visualize with color
plt.scatter(MYO_reduc_df['x'], MYO_reduc_df['y'], c=labels)
plt.show()
```

# Using K-means to create a scatter plot and elbow

```python
# Finding the best value for k
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(MYO_reduc_df)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('MYOPIA Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
df_elbow
```

```python
# Predicting clusters 

# Initialize the k-means model
model = KMeans(n_clusters=5, random_state=0)

# Fit the model
model.fit(MYO_reduc_df)

# Predict clusters
predictions = model.predict(MYO_reduc_df)

# Add the predicted class column to the dataframe
MYO_reduc_df["MYOPIA"] = model.labels_
MYO_reduc_df.head()
```

```python
plt.scatter(x=MYO_reduc_df['x'], y=MYO_reduc_df['y'], c=MYO_reduc_df['MYOPIA'])
plt.xlabel('X')
plt.ylabel('y')
plt.title('MYOPIA clusters')
plt.show()
```
